```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt


def f(x, u, p=2, k=1):
    return 1 / (1 + ((x-u)/k)**p)


x = np.arange(0.0, 10.0, 0.01)
y = f(x, 5)
fig, ax = plt.subplots()
ax.plot(x, y, color='0')

y2 = f(x, 0, p=4, k=3)
ax2 = ax.twinx()
ax2.plot(x, y2, color='0')

y3 = f(x, 10, p=4, k=3)
ax3 = ax.twinx()
ax3.plot(x, y3, color='0')

ax.set(xlabel='x', ylabel=r'$\mu$')  # title=u'Текст на русскм')

ax.grid()

fig.savefig('img/fig4.png', dpi=300)
```
