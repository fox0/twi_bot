```python
import matplotlib.pyplot as plt
import networkx as nx

node_size = 100

graph = nx.Graph()
for i in range(6):
    graph.add_node(i)
    graph.add_edge(1, i)
plt.subplot(221)
nx.draw(graph, node_color='0', node_size=node_size)

graph = nx.Graph()
for i in range(6):
    graph.add_node(i)
    for j in range(6):
        graph.add_edge(i, j)
plt.subplot(222)
nx.draw_circular(graph, node_color='0', node_size=node_size)

graph = nx.Graph()
for i in range(6):
    graph.add_node(i)
    graph.add_edge(1, i)
graph.add_node(6)
graph.add_node(7)
graph.add_node(8)
graph.add_node(9)
graph.add_edge(5, 6)
graph.add_edge(5, 7)
graph.add_edge(5, 8)
graph.add_edge(4, 9)

plt.subplot(212)
nx.draw(graph, node_color='0', node_size=node_size)

plt.savefig('fi2.png', dpi=300)
```

```python
# coding: utf-8
import matplotlib.pyplot as plt
import numpy as np


def f(x, a):
    return 1 / (1 + (x-a)**2)


x = np.arange(0.0, 10.0, 0.01)
y = f(x, 5)
y2 = f(x, 3)


fig, ax = plt.subplots()
ax.plot(x, y, color='0')

ax2 = ax.twinx()

# ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(x, y2, color='0')
# ax2.tick_params(axis='y', labelcolor=color)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title=u'  ')
ax.grid()

fig.savefig('fig3.png', dpi=300)
# plt.show()
```
