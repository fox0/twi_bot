```python
import matplotlib.pyplot as plt
import networkx as nx

node_size = 100

graph = nx.Graph()
for i in range(6):
    graph.add_node(i)
    graph.add_edge(1, i)
plt.subplot(221)
nx.draw(graph, node_color='0', node_size=node_size)

graph = nx.Graph()
for i in range(6):
    graph.add_node(i)
    for j in range(6):
        graph.add_edge(i, j)
plt.subplot(222)
nx.draw_circular(graph, node_color='0', node_size=node_size)

graph = nx.Graph()
for i in range(6):
    graph.add_node(i)
    graph.add_edge(1, i)
graph.add_node(6)
graph.add_node(7)
graph.add_node(8)
graph.add_node(9)
graph.add_edge(5, 6)
graph.add_edge(5, 7)
graph.add_edge(5, 8)
graph.add_edge(4, 9)

plt.subplot(212)
nx.draw(graph, node_color='0', node_size=node_size)

plt.savefig('img/fig1.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

graph = nx.Graph()
for i in range(4):
    for j in range(i*10, i*10+10):
        graph.add_edge(i*10+1, j)
    graph.add_edge(i*10+1, 100)
nx.draw_spring(graph, node_color='0', node_size=100)

plt.savefig('img/fig2.png', dpi=300)
```
